```python
%%manim -qm -v WARNING numpyBroadcasting

import numpy as np
import torch
import torch.nn.functional as F
class numpyBroadcasting(Scene):
  def construct(self):
    # create numpy arrays
    A = np.round(np.array([[1,1],[2,2],[3,3]], dtype=np.float32),2)
    A1 = A[:,1].reshape(A.shape[0],1)
    B = np.round(np.array([[10,10],[20,20],[30,30]], dtype=np.float32),2)
    B1 = B[:,1].reshape(1,B.shape[0])
    C = np.round(np.sqrt(np.square(A[:, None, :] - B[None, :, :]).sum(axis=-1)),1)

    fontsize = 100
    scale = 0.7
    ############################################################################
    matrixA = MathTable(A,include_outer_lines=True)
    matrixA.color = BLUE_B
    matrixA.scale(scale)
    ############################################################################
    matrixA1 = MathTable(A1,include_outer_lines=True)
    matrixA1.color = BLUE_B
    matrixA1_1 =  MathTable(A1,include_outer_lines=True)
    matrixA1_1.color = BLUE_B
    matrixA1_1.shift(0.2 * DOWN)
    matrixA1_1.shift(0.15 * LEFT)
    matrixA1_group = VGroup(matrixA1, matrixA1_1).scale(scale)
    ############################################################################
    matrixB = MathTable(B,include_outer_lines=True)
    matrixB.color = BLUE_B
    matrixB.scale(scale)
    ############################################################################
    matrixB1 = MathTable(B1,include_outer_lines=True)
    matrixB1.color = BLUE_B
    matrixB1_1 =  MathTable(B1,include_outer_lines=True)
    matrixB1_1.color = BLUE_B
    matrixB1_1.shift(0.2 * UP)
    matrixB1_1.shift(0.1 * RIGHT)
    matrixB1_group = VGroup(matrixB1, matrixB1_1).scale(scale)
    ############################################################################
    matrixC = MathTable(C,include_outer_lines=True)
    matrixC.color = BLUE_B
    matrixC.scale(scale)
    ############################################################################
    code = '''
      C = np.sqrt(np.square(A[:, None, :] - B[None, :, :]).sum(axis=-1))
        '''
    CodeC = Code(code=code, tab_width=1, background="rectangle",
                    language="Python", font="Monospace").scale(0.8)
    ############################################################################
    text = Text('https://mashaan14.github.io/mashaan/manim').scale(0.4)
    text.to_edge(DOWN+LEFT)
    self.play(Write(text))

    matrixA.to_edge(LEFT)
    matrixA1_group.to_edge(LEFT)
    matrixB.to_edge(RIGHT)
    matrixB1_group.to_edge(RIGHT)
    self.play(Write(matrixA),Write(matrixB))

    CodeC.to_edge(UP+LEFT)
    self.play(FadeIn(CodeC))

    self.wait(2)

    self.play(TransformMatchingShapes(matrixA, matrixA1_group),TransformMatchingShapes(matrixB, matrixB1_group))

    self.play(matrixA1_group.animate.next_to(matrixC, LEFT))
    self.play(matrixB1_group.animate.next_to(matrixC, UP))
    self.play(Write(matrixC))

    self.wait(5)
```

```python
# verify the distances are correct using scipy distance function
from scipy.spatial import distance
A = np.array([[1,1],[2,2],[3,3]])
B = np.array([[10,10],[20,20],[30,30]])
distance.cdist(A, B, 'euclidean')
```
