```python
%%manim -qm -v WARNING visualizeSoftmax

import numpy as np
import torch
import torch.nn.functional as F
class visualizeSoftmax(Scene):
  def construct(self):
    A = np.array([[1, 1, 1], [2, 2, 2], [3, 3, 3], [4, 4, 4]], dtype=np.float32)
    B_tensor = F.softmax(torch.from_numpy(A),dim=0)
    B = np.round(B_tensor.cpu().detach().numpy(),4)
    C_tensor = F.softmax(torch.from_numpy(A),dim=1)
    C = np.round(C_tensor.cpu().detach().numpy(),4)

    fontsize = 100
    ############################################################################
    matrixA = MathTable(A,include_outer_lines=True)
    matrixA.height = 4
    matrixA.color = BLUE_B
    textA = Text('MLP output', color=BLUE_B)
    matrixA_group = VGroup(matrixA,
                  textA.next_to(matrixA, UP)).scale(0.6)
    ############################################################################
    matrixB = MathTable(B,include_outer_lines=True)
    matrixB.color = BLUE_B
    code = '''
          import torch.nn.functional as F
          F.softmax(A,dim=0)
            '''
    CodeB = Code(code=code, tab_width=1, background="rectangle",
                        language="Python", font="Monospace")
    matrixB_group = VGroup(matrixB, CodeB.next_to(matrixB, UP)).scale(0.6)
    ############################################################################
    matrixC = MathTable(C,include_outer_lines=True)
    matrixC.color = BLUE_B
    code = '''
          import torch.nn.functional as F
          F.softmax(A,dim=1)
            '''
    CodeC = Code(code=code, tab_width=1, background="rectangle",
                        language="Python", font="Monospace")
    matrixC_group = VGroup(matrixC, CodeC.next_to(matrixC, UP)).scale(0.6)
    ############################################################################
    stuff = VGroup(matrixA_group,matrixB_group,matrixC_group)
    stuff.arrange_in_grid(1, 3, row_alignments='d')

    text = Text('https://mashaan14.github.io/mashaan/manim').scale(0.4)
    text.to_edge(DOWN+LEFT)
    self.play(Write(text))

    self.play(Write(stuff[0]))

    self.play(Write(stuff[1]))
    textB = Text('When dim=0 all columns sum to 1',t2c={"dim=0": YELLOW}).scale(0.5)
    textB.to_edge(UP+LEFT)
    self.play(FadeIn(textB))
    for i in range(len(matrixB.get_columns())):
      self.play(Indicate(matrixB.get_columns()[i]))
    self.play(FadeOut(textB))

    self.play(Write(stuff[2]))
    textC = Text('When dim=1 all rows sum to 1',t2c={"dim=1": YELLOW}).scale(0.5)
    textC.to_edge(UP+LEFT)
    self.play(FadeIn(textC))
    for i in range(len(matrixC.get_rows())):
      self.play(Indicate(matrixC.get_rows()[i]))
    self.play(FadeOut(textC))
    self.wait()
```
